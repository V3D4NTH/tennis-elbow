```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import joblib

# Load feature CSVs
correct_df = pd.read_csv("/content/updated_yolo_correct_labels/correct_output.csv")
correct_df = correct_df.drop("Frame", axis=1)
incorrect_df = pd.read_csv("/content/updated_yolo_incorrect_labels/incorrect_output.csv")
incorrect_df = incorrect_df.drop("Frame", axis=1)

# Add labels
correct_df['label'] = 1
incorrect_df['label'] = 0

# Combine datasets
df = pd.concat([correct_df, incorrect_df], ignore_index=True)

# Separate features and label
X = df.drop(columns=["label"])
y = df["label"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Train Random Forest classifier
clf = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))

# Save model and scaler for inference
joblib.dump(clf, "pose_classifier_rf.pkl")
joblib.dump(scaler, "pose_scaler.pkl")

print("\n✅ Model and scaler saved successfully.")
```

```python
import cv2
import numpy as np
import joblib
from ultralytics import YOLO
from sklearn.preprocessing import StandardScaler

# Load model and scaler
clf = joblib.load("pose_classifier_rf.pkl")
scaler = joblib.load("pose_scaler.pkl")

# Load YOLOv11 pose model
yolo_model = YOLO("yolo11x-pose.pt")

# Input video
input_path = "/content/first_exercise/incorrect_Exercises/first_exercise_incorrect_2.mp4"
cap = cv2.VideoCapture(input_path)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("inference_output5.mp4", fourcc, fps, (width, height))

# Helper: calculate angle between 3 points
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    ba = a - b
    bc = c - b
    cosine = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-6)
    angle = np.arccos(np.clip(cosine, -1.0, 1.0))
    return np.degrees(angle)

# Track previous angles for derivatives
prev_elbow, prev_prev_elbow = None, None
prev_wrist, prev_wrist_point = None, None

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = yolo_model.predict(frame, conf=0.3, verbose=False)
    if results and results[0].keypoints and results[0].keypoints.xy is not None:
        keypoints = results[0].keypoints.xy.cpu().numpy()
        if keypoints.shape[0] == 0:
            out.write(frame)
            continue

        person = keypoints[0]
        try:
            # Angles
            left_shoulder = calculate_angle(person[3], person[5], person[7])
            left_elbow = calculate_angle(person[5], person[7], person[9])
            left_wrist = calculate_angle(person[7], person[9], person[9] + [0.001, 0])

            # Deltas
            elbow_delta = left_elbow - prev_elbow if prev_elbow is not None else 0
            elbow_acc = (left_elbow - 2 * prev_elbow + prev_prev_elbow
                         if prev_elbow is not None and prev_prev_elbow is not None else 0)
            wrist_delta = left_wrist - prev_wrist if prev_wrist is not None else 0
            wrist_disp = np.linalg.norm(person[9] - prev_wrist_point) if prev_wrist_point is not None else 0

            prev_prev_elbow = prev_elbow
            prev_elbow = left_elbow
            prev_wrist = left_wrist
            prev_wrist_point = person[9]

            # Feature vector
            features = np.array([[left_shoulder, left_elbow, left_wrist,
                                  elbow_delta, elbow_acc, wrist_delta, wrist_disp]])
            features_scaled = scaler.transform(features)

            # Predict
            pred = clf.predict(features_scaled)[0]
            label = "Correct" if pred == 1 else "Incorrect"
            color = (0, 255, 0) if pred == 1 else (0, 0, 255)

            # Display label
            cv2.putText(frame, f"Prediction: {label}", (30, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2, cv2.LINE_AA)

        except Exception as e:
            pass

    out.write(frame)

cap.release()
out.release()
print("✅ Inference complete. Saved as inference_output.mp4")
```
